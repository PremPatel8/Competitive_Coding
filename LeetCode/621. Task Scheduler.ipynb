## 621. Task Scheduler
- Description:
  <blockquote>
    You are given an array of CPU `tasks`, each labeled with a letter from A to Z, and a number `n`. Each CPU interval can be idle or allow the completion of one task. Tasks can be completed in any order, but there's a constraint: there has to be a gap of **at least** `n` intervals between two tasks with the same label.

    Return the **minimum** number of CPU intervals required to complete all tasks.

    **Example 1:**

    **Input:** tasks = \["A","A","A","B","B","B"\], n = 2

    **Output:** 8

    **Explanation:** A possible sequence is: A -> B -> idle -> A -> B -> idle -> A -> B.

    After completing task A, you must wait two intervals before doing A again. The same applies to task B. In the 3<sup>rd</sup> interval, neither A nor B can be done, so you idle. By the 4<sup>th</sup> interval, you can do A again as 2 intervals have passed.

    **Example 2:**

    **Input:** tasks = \["A","C","A","B","D","B"\], n = 1

    **Output:** 6

    **Explanation:** A possible sequence is: A -> B -> C -> D -> A -> B.

    With a cooling interval of 1, you can repeat a task after just one other task.

    **Example 3:**

    **Input:** tasks = \["A","A","A", "B","B","B"\], n = 3

    **Output:** 10

    **Explanation:** A possible sequence is: A -> B -> idle -> idle -> A -> B -> idle -> idle -> A -> B.

    There are only two types of tasks, A and B, which need to be separated by 3 intervals. This leads to idling twice between repetitions of these tasks.

    **Constraints:**

    -   `1 <= tasks.length <= 10<sup>4</sup>`
    -   `tasks[i]` is an uppercase English letter.
    -   `0 <= n <= 100`
  </blockquote>

### Solution 1, Greedy Approach Using Math Formula

### ðŸ”‘ Key Idea: Schedule the Most Frequent Task First

Suppose the most frequent task (say `'A'`) appears `max_val` times.

To respect the cooldown `n`, you need **at least `n` other tasks or idle slots** between two `'A'`s.

So, you can think of building a **framework** like this:

```
A _ _ _ A _ _ _ A _ _ _ A        â† if max_val = 4, n = 3
```
-   There are `max_val - 1` **gaps** (or "parts") between the `A`s.
-   Each gap must be **at least `n` slots long** â†’ so each block (including the `A`) is `n + 1` units:
    -   `[A _ _ _] [A _ _ _] [A _ _ _] A`
-   The last `A` doesnâ€™t need a cooldown after it.

So, the **skeleton** takes:
(max_val - 1) * (n + 1) slots
Then add the final A â†’ but wait! What if multiple tasks tie for max_val?

### ðŸ”„ Handling Ties: `max_count`

Suppose both `'A'` and `'B'` appear `max_val = 4` times.

Then your schedule must end with **all** of them:

```
A B _ _ A B _ _ A B _ _ A B
```
So the last block has **`max_count` tasks**, not just 1.

Thus, total length becomes:

> `(max_val - 1) * (n + 1) + max_count`

âœ… This accounts for:

-   `max_val - 1` full cycles (each of size `n + 1`)
-   Plus `max_count` tasks at the end (all the most frequent ones)

### âš–ï¸ But What If There Are Enough Other Tasks?

If you have **so many other tasks** that they **fill all idle slots**, then you donâ€™t need any idles â€” total time is just `len(tasks)`.

Example:  
Tasks = `["A","A","B","B","C","C"]`, `n = 1`  
â†’ You can do: `A B C A B C` â†’ no idles needed.

So final answer is:

```python
max(len(tasks), (max_val - 1) * (n + 1) + max_count)
```

---

Let N be the number of tasks.
- Time Complexity: O(N)
  - The loop iterating over the tasks array has a time complexity of O(N). The loop iterating over the freq array has a time complexity proportional to the number of unique tasks, which is at most 26 because the tasks are represented by the letters A to Z. Therefore, the overall time complexity is O(N+26), which simplifies to O(N).
- Space Complexity: O(26) = O(1)
  - The freq array can store at most 26 unique tasks, resulting in O(26) space complexity. Other variables used in the algorithm have constant space requirements. Therefore, the overall space complexity is O(1).

````python
class Solution:
    def leastInterval(self, tasks: List[str], n: int) -> int:
        task_freq = Counter(tasks)
        max_freq = max(task_freq.values())
        max_freq_count = sum(1 for count in task_freq.values() if count == max_freq)
        
        # Calculate minimum intervals required
        min_intervals = (max_freq - 1) * (n + 1) + max_freq_count
        
        # The answer is the maximum between total tasks and computed idle-aware schedule
        return max(len(tasks), min_intervals)
````

### Solution 2, Filling the Slots and Sorting
Let the number of tasks be N
- Time Complexity: O(N)
  - The time complexity of the algorithm is O(26log26+N), where 26log26 is the time complexity of sorting the frequency array, and N is the length of the input task list, which is the dominating term.
- Space Complexity: O(26)=O(1)
  - The frequency array has a size of 26.
  - Note that some extra space is used when we sort arrays in place. The space complexity of the sorting algorithm depends on the programming language.

    - In Python, the sort method sorts a list using the Timsort algorithm which is a combination of Merge Sort and Insertion Sort and has O(N) additional space.
    - In Java, Arrays.sort() is implemented using a variant of the Quick Sort algorithm which has a space complexity of O(logN) for sorting array.
    - In C++, the sort() function is implemented as a hybrid of Quick Sort, Heap Sort, and Insertion Sort, with a worse-case space complexity of O(logN).

We sort the frequency array, which has a size of 26. The space used for sorting takes O(26) or O(log26), which is constant, so the space complexity of the algorithm is O(26), which is constant, i.e. O(1).

````python
class Solution:
    def leastInterval(self, tasks: List[str], n: int) -> int:
        # Create a frequency array to keep track of the count of each task
        freq = [0] * 26
        for task in tasks:
            freq[ord(task) - ord('A')] += 1

        # Sort the frequency array in non-decreasing order
        freq.sort()
        # Calculate the maximum frequency of any task
        # Subtract 1 because we want to find the number of intervals, not the number of occurrences.
        max_freq = freq[25] - 1
        # Calculate the number of idle slots that will be required
        idle_slots = max_freq * n
        # Iterate over the frequency array from the second highest frequency to the lowest frequency


        #The loop starts from the second highest frequency and goes down to the lowest frequency
        # This ensures that the highest frequency task's idle slots are considered only once, 
        # as it was accounted for when calculating maxFreq in the earlier step.
        for i in range(24, -1, -1):
            # Subtract the minimum of the maximum frequency and the current frequency from the idle slots
            # For any other task with frequency f, you can place at most one instance per gap, 
            # and only up to max_freq times, because there are only max_freq gaps.
            # Any extra occurrences (beyond max_freq) will be scheduled after the last main task, and do not reduce idle time.
            idle_slots -= min(max_freq, freq[i])

        # After filling the idle slots in the above loop, 
        # if there are remaining idle slots, they must be included in the total time
        # If no idle slots remain, then all tasks fit perfectly â€” total time is just the number of tasks.
        # idle_slots can become negative during the loop â€” that just means we have more tasks than idle slots, so no idling is needed.
        return idle_slots + len(tasks) if idle_slots > 0 else len(tasks)
        
````

### Solution 3, Slowest, Greedy + Priority Queue / Max Heap Simulation
In Python 3, frequencies are stored as negative values to simulate a max-heap behavior.

Let the number of tasks be N. Let k be the size of the priority queue. k can, at maximum, be 26 because the priority queue stores the frequency of each distinct task, which is represented by the letters A to Z.

- Time Complexity: O(N)
  - In the worst case, all tasks must be processed, and each task might be inserted and extracted from the priority queue. 
  - The priority queue operations (insertion and extraction) have a time complexity of O(logk) each. Therefore, the overall time complexity is O(Nâ‹…logk). 
  - Since k is at maximum 26, logk is a constant term. We can simplify the time complexity to O(N). This is a linear time complexity with a high constant factor.
- Space Complexity: O(26) = O(1)
  - The space complexity is mainly determined by the frequency array and the priority queue. 
  - The frequency array has a constant size of 26, and the priority queue can have a maximum size of 26 when all distinct tasks are present. 
  - Therefore, the overall space complexity is O(1) or O(26), which is considered constant.

````python
class Solution:
    def leastInterval(self, tasks: List[str], n: int) -> int:
        # Build frequency map
        
        # More general and readable. Works for any hashable task label (not just 'A'â€“'Z').
        # Slight overhead due to dictionary creation, but negligible for typical inputs.
        task_freq = Counter(tasks)
        
        """
        # Alt approach, Slightly more efficient in time and space if the input constraint guarantees only uppercase letters.
        task_freq = [0] * 26
        
        for ch in tasks:
            task_freq[ord(ch) - ord('A')] += 1
        """
        
        max_heap = [-freq for freq in task_freq.values() if freq > 0]
        heapq.heapify(max_heap)
        
        intervals = 0

        while max_heap:
            cycle = n + 1
            temp_store = []
            task_count = 0
            
            # Execute tasks in each cycle
            while cycle > 0 and max_heap:
                curr_freq = -heapq.heappop(max_heap)
                if curr_freq > 1:
                    temp_store.append(-(curr_freq - 1))
                task_count += 1
                cycle -= 1
            
            # Restore updated frequencies to the heap
            for updated_freq in temp_store:
                heapq.heappush(max_heap, updated_freq)
            
            # Add time for the completed cycle
            intervals += task_count if not max_heap else n + 1
            
        return intervals
````
