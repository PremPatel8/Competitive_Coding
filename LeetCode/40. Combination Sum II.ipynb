## 40. Combination Sum II
- Description:
  <blockquote>
    [40. Combination Sum II](https://leetcode.com/problems/combination-sum-ii/) Given a collection of candidate numbers (`candidates`) and a target number (`target`), find all unique combinations in `candidates` where the candidate numbers sum to `target`.
     
    Each number in `candidates` may only be used **once** in the combination.
     
    **Note:** The solution set must not contain duplicate combinations.
     
    **Example 1:**
    **Input:** candidates = [10,1,2,7,6,1,5], target = 8
    **Output:** 
    [
    [1,1,6],
    [1,2,5],
    [1,7],
    [2,6]
    ]
     
    **Example 2:**
    **Input:** candidates = [2,5,2,1,2], target = 5
    **Output:** 
    [
    [1,2,2],
    [5]
    ]
     
    **Constraints:**
     
    - `1 <= candidates.length <= 100`
    - `1 <= candidates[i] <= 50`
    - `1 <= target <= 30`
  </blockquote>

### Solution 1, Backtracking
Solution description
- Time Complexity: O(2^N)
  - In the worst case (no duplicates, small numbers), you might explore every subset (each element: pick or not pick) → 2ⁿ subsets.
- Space Complexity: O(N)

```python
class Solution:
    def combinationSum2(self, candidates: List[int], target: int) -> List[List[int]]:
        def backtrack(result, candidates, remaining, curr_comb, start):
            if remaining < 0:
                return
            elif remaining == 0:
                result.append(curr_comb.copy())
            else:
                for idx in range(start, len(candidates)):
                    if idx > start and candidates[idx] == candidates[idx - 1]:
                        continue
                    
                    if remaining < candidates[idx]:
                        break
                    
                    curr_comb.append(candidates[idx])
                    backtrack(result, candidates, remaining - candidates[idx], curr_comb, idx + 1)
                    curr_comb.pop()

        result = []
        candidates.sort()
        backtrack(result, candidates, target, [], 0)
        return result
```
