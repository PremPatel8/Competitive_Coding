## 9. Palindrome Number
- Description:
  <blockquote>
    Given an integer x, return true if x is a , and false otherwise.

    

    Example 1:

    Input: x = 121
    Output: true
    Explanation: 121 reads as 121 from left to right and from right to left.

    Example 2:

    Input: x = -121
    Output: false
    Explanation: From left to right, it reads -121. From right to left, it becomes 121-. Therefore it is not a palindrome.

    Example 3:

    Input: x = 10
    Output: false
    Explanation: Reads 01 from right to left. Therefore it is not a palindrome.

    

    Constraints:

        -231 <= x <= 231 - 1

    
    Follow up: Could you solve it without converting the integer to a string?
  </blockquote>

### Solution 1, Revert back half of the number and check if it matches with the first half
Solution description
- Time Complexity: O(log10​(n))
  - We divided the input by 10 for every iteration, so the time complexity is O(log10​(n)), log N to the base 10
- Space Complexity: O(1)

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        # Special cases:
        # When x < 0, x is not a palindrome.
        # Also if the last digit of the number is 0, in order to be a palindrome,
        # the first digit of the number also needs to be 0.
        # Only 0 satisfy this property.
        if x < 0 or (x % 10 == 0 and x != 0):
            return False

        revertedNumber = 0
        while x > revertedNumber:
            # x % 10 - Get the last digit of the number x
            revertedNumber = revertedNumber * 10 + x % 10

            # x // 10 - Remove the last digit of the number x
            x = x // 10

        # When the length is an odd number, we can get rid of the middle digit by revertedNumber//10
        # For example when the input is 12321, at the end of the while loop we get x = 12, revertedNumber = 123,
        # since the middle digit doesn't matter in palidrome(it will always equal to itself), we can simply get rid of it.
        return x == revertedNumber or x == revertedNumber // 10
```

### Solution 2, Convert to string and use two pointers from left and right boundary to check matching characters
Solution description
- Time Complexity: O(N)
- Space Complexity: O(N)

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        noStr = str(x)
        i = 0
        j = len(noStr)-1

        while i < j:
            if noStr[i] != noStr[j]:
                return False
            
            i += 1
            j -= 1
            
        
        return True
```

### Solution 3, Convert number to string, reverse entire string and compare to original
Solution description
- Time Complexity: O(N)
- Space Complexity: O(N)

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        s = str(x)
        return s == s[::-1]
```
