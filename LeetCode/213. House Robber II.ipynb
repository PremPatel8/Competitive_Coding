## 213. House Robber II
- Description:
  <blockquote>
    You are a professional robber planning to rob houses along a street. Each house has a certain amount of money stashed. All houses at this place are **arranged in a circle.** That means the first house is the neighbor of the last one. Meanwhile, adjacent houses have a security system connected, andÂ **it will automatically contact the police if two adjacent houses were broken into on the same night**.

  Given an integer array `nums` representing the amount of money of each house, return _the maximum amount of money you can rob tonight **without alerting the police**_.

  **Example 1:**

  ```
  Input: nums = [2,3,2]
  Output: 3
  Explanation: You cannot rob house 1 (money = 2) and then rob house 3 (money = 2), because they are adjacent houses.

  ```

  **Example 2:**

  ```
  Input: nums = [1,2,3,1]
  Output: 4
  Explanation: Rob house 1 (money = 1) and then rob house 3 (money = 3).
  Total amount you can rob = 1 + 3 = 4.

  ```

  **Example 3:**

  ```
  Input: nums = [1,2,3]
  Output: 3

  ```

  **Constraints:**

  -   `1 <= nums.length <= 100`
  -   `0 <= nums[i] <= 1000`
  </blockquote>

### Solution 1, Recursion with Memoization
Solution description
- Time Complexity: O(N)
- Space Complexity: O(N)

````python
class Solution:
    def rob(self, nums: List[int]) -> int:
        numsLen = len(nums)
        
        if numsLen == 1:
            return nums[0]

        def robFrom(idx, terminal_idx, memo):
            if idx > terminal_idx:
                return 0

            if idx in memo:
                return memo[idx]
            
            ans = max(nums[idx]+robFrom(idx+2, terminal_idx, memo), robFrom(idx+1, terminal_idx, memo))

            memo[idx] = ans

            return ans
        
        return max(robFrom(0, numsLen-2, {}), robFrom(1, numsLen-1, {}))
````

### Solution 2, Dynamic Programming
Solution description
- Time Complexity: O(N)
- Space Complexity: O(N)

````python
class Solution:
    def rob(self, nums: List[int]) -> int:
        numlen = len(nums)

        if numlen == 1:
            return nums[0]
        
        if numlen == 2:
            return max(nums[0], nums[1])


        def rob_sublist(nums_sublist):
            sublist_len = len(nums_sublist)
            
            dp = [0] * sublist_len
            dp[0] = nums_sublist[0]
            dp[1] = max(nums_sublist[0], nums_sublist[1])

            for idx in range(2, sublist_len):
                dp[idx] = max(nums_sublist[idx]+dp[idx-2], dp[idx-1])
            
            return dp[sublist_len-1]
        
        return max(rob_sublist(nums[0:numlen-1]), rob_sublist(nums[1:numlen]))
````

### Solution 3, Space Optimized Dynamic Programming
Solution description
- Time Complexity: O(N)
- Space Complexity: O(1)
  - since we are not consuming additional space other than variables for two previous results and a temporary variable to hold one of the previous results.

````python
class Solution:
    def rob(self, nums: List[int]) -> int:
        if nums is None or len(nums) == 0:
            return 0

        numlen = len(nums)

        if numlen == 1:
            return nums[0]
        
        if numlen == 2:
            return max(nums[0], nums[1])

        return max(self.rob_sublist(nums[0:numlen-1]), self.rob_sublist(nums[1:numlen]))

    def rob_sublist(self, nums_sublist):
            prevTwo = 0
            prevOne = 0

            for no in nums_sublist:
                cur = max(no+prevTwo, prevOne)
                prevTwo = prevOne
                prevOne = cur
            
            return cur
````
