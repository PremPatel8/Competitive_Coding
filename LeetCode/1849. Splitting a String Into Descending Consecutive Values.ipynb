## 1849. Splitting a String Into Descending Consecutive Values
- Description:
  <blockquote>
  [1849. Splitting a String Into Descending Consecutive Values](https://leetcode.com/problems/splitting-a-string-into-descending-consecutive-values/) You are given a string `s` that consists of only digits.
 
Check if we can split `s` into **two or more non-empty substrings** such that the **numerical values** of the substrings are in **descending order** and the **difference** between numerical values of every two **adjacent** **substrings** is equal to `1`.
 
- For example, the string `s = "0090089"` can be split into `["0090", "089"]` with numerical values `[90,89]`. The values are in descending order and adjacent values differ by `1`, so this way is valid.
- Another example, the string `s = "001"` can be split into `["0", "01"]`, `["00", "1"]`, or `["0", "0", "1"]`. However all the ways are invalid because they have numerical values `[0,1]`, `[0,1]`, and `[0,0,1]` respectively, all of which are not in descending order.
 
Return `true`  *if it is possible to split*  `s`​​​​​​  *as described above*  *, or* `false` *otherwise.*
 
A **substring** is a contiguous sequence of characters in a string.
 
**Example 1:**
**Input:** s = "1234"
**Output:** false
**Explanation:** There is no valid way to split s.
 
**Example 2:**
**Input:** s = "050043"
**Output:** true
**Explanation:** s can be split into ["05", "004", "3"] with numerical values [5,4,3].
The values are in descending order with adjacent values differing by 1.
 
**Example 3:**
**Input:** s = "9080701"
**Output:** false
**Explanation:** There is no valid way to split s.
 
**Constraints:**
 
- `1 <= s.length <= 20`
- `s` only consists of digits.
  </blockquote>

### Solution 1, Implicit Backtracking (no parts list) with Early Termination

- Time: O(n²) — unchanged
The pruning and traversal logic is identical, so the asymptotic bound stays the same.

- Space: O(n) — same bound, but better constant
Asymptotically the same, but the removal of the parts list cuts the memory usage roughly in half in practice.

````python
class Solution:
    def splitString(self, s: str) -> bool:
        if len(s) <= 1:
            return False

        strlen = len(s)

        def backtrack(prev_no, parts_count, idx):
            if idx == strlen:
                return parts_count >= 2

            for i in range(idx, strlen):
                num = int(s[idx : i+1])

                if prev_no is None or num == prev_no - 1:
                    if backtrack(num, parts_count + 1, i + 1):
                        return True
                elif num > prev_no - 1:
                    break  # monotonic pruning

            return False

        return backtrack(None, 0, 0)
````

### Solution 2, Backtracking with with parts list and Early Termination

Building a parts list that contains the valid substrings in case the future requirements change such that we are expected to return this list instead of just True / False response

- Time: O(n²) — same asymptotic bound as before

    The pruning doesn't change the worst case, but improves the constant factor:

    O(n) choices for the first number (no target constraint)
    For each first number, the pruning means subsequent levels scan only until num > target then break, rather than scanning to the end — but it's still O(n) total scan per branch in the worst case

- Space: O(n) — unchanged

    Recursion depth: O(n)
    parts list: O(n)

    The real win from pruning is practical performance, not asymptotic complexity. On strings where the target value is exceeded early, many iterations are skipped entirely.


````python
class Solution:
    def splitString(self, s: str) -> bool:
        strlen = len(s)

        if len(s) <= 1:
            return False

        def backtrack(parts, idx):
            if idx == strlen:
                return len(parts) >= 2
            
            for i in range(idx, strlen):
                num = int(s[idx : i+1])
                target = parts[-1] - 1 if parts else None

                if target is None or num == target:
                    parts.append(num)
                    if backtrack(parts, i+1):
                        return True
                    parts.pop()
                elif num > target:
                    break  # ← pruning: num will only grow larger
            
            return False
        
        return backtrack([], 0)
````

### Solution 3, Backtracking with parts list solution

## Time Complexity: **O(n²)**

Here's the breakdown:

- **First number:** O(n) choices — it can span length 1 to n
- **Subsequent numbers:** once a value is chosen, the next target is fixed (`prev - 1`), so branching is heavily constrained — at most O(n) characters are scanned at each level to find the match
- **Recursion depth:** at most O(n) levels (e.g., `"321"` splits into n single-digit parts)

So the dominant cost is:

```
O(n) first choices × O(n) scan per level = O(n²)
```

> The `num == parts[-1] - 1` constraint is what keeps this from blowing up to O(2ⁿ) — it eliminates nearly all branches after the first number is chosen.

---

## Space Complexity: **O(n)**

Two contributors:

| Factor | Cost |
|---|---|
| Recursion call stack depth | O(n) |
| `parts` list at any point | O(n) |

Both are bounded by n since the maximum number of parts equals the string length (all single-digit splits).

---

- Time Complexity: O(n²)
- Space Complexity: O(N)

````python
class Solution:
    def splitString(self, s: str) -> bool:
        strlen = len(s)

        if len(s) <= 1:
            return False

        def backtrack(parts, idx):
            if idx == strlen:
                return len(parts) >= 2
            
            for i in range(idx, strlen):
                num = int(s[idx : i+1])

                if not parts or num == parts[-1] - 1:
                    parts.append(num)
                    if backtrack(parts, i+1):
                        return True
                    parts.pop()
            
            return False
        
        return backtrack([], 0)
````
