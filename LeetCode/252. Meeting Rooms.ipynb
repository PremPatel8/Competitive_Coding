## 252. Meeting Rooms
- Description:
  <blockquote>
    Given an array of meeting time `intervals` where `intervals[i] = [starti, endi]`, determine if a person could attend all meetings.
     
    **Example 1:**
    **Input:** intervals = `[0, 30],[5,10],[15,20]`
    **Output:** false
     
    **Example 2:**
    **Input:** intervals = `[7, 10],[2,4]`
    **Output:** true
     
    **Constraints:**
     
    - `0 <= intervals.length <= 104`
    - `intervals[i].length == 2`
    - `0 <= starti< endi<= 106`
  </blockquote>

- Difficulty: Easy

### Solution 1, Greedy Approach: Sorting by Start Time
Solution description

- Time Complexity: O(NlogN)
- Space Complexity: O(1) or O(N) depending on the default sorting algo of the programming language

```python
class Solution:
    def canAttendMeetings(self, intervals: List[List[int]]) -> bool:
        if len(intervals) == 0:
            return True

        # Sort meeting intervals by start time
        intervals.sort(key = lambda X:X[0])
        
        for i in range(1, len(intervals)):
            # If the current meeting starts before the previous one ends, we found an overlap
            if intervals[i][0] < intervals[i-1][1]:
                return False
                
        # If no overlaps are found
        return True
```
