## 560. Subarray Sum Equals K
- Description:
  <blockquote>
    Given an array of integers nums and an integer k, return the total number of subarrays whose sum equals to k.

    A subarray is a contiguous non-empty sequence of elements within an array.

    

    Example 1:

    Input: nums = [1,1,1], k = 2
    Output: 2

    Example 2:

    Input: nums = [1,2,3], k = 3
    Output: 2

    

    Constraints:

        1 <= nums.length <= 2 * 104
        -1000 <= nums[i] <= 1000
        -107 <= k <= 107


  </blockquote>

### Solution 1, Prefix Sum with HashMap
Solution description
- Time Complexity: O(N)
  - single pass
- Space Complexity: O(N)
  - HashMap storage

```python
class Solution:
    def subarraySum(self, nums: List[int], k: int) -> int:
        count = 0
        prefix_sum = 0
        sum_freq = {0: 1}  # Initialize with 0 sum seen once
        
        for num in nums:
            prefix_sum += num
            
            # Check if (prefix_sum - k) exists in hashmap
            if (prefix_sum - k) in sum_freq:
                count += sum_freq[prefix_sum - k]
            
            # Add current prefix_sum to hashmap
            sum_freq[prefix_sum] = sum_freq.get(prefix_sum, 0) + 1
        
        return count
```

### Solution 2, TLE, Space Efficient without Prefix Sum array
Solution description
Time Complexity: O(n²) - checking all possible subarrays  
Space Complexity: O(1)

```python
class Solution:
    def subarraySum(self, nums: List[int], k: int) -> int:
        count = 0
        for start in range(len(nums)):
            sum = 0
            for end in range(start, len(nums)):
                sum += nums[end]
                if sum == k:
                    count += 1
        return count
```

### Solution 2, TLE, Naive Prefix Sum sol

Time Complexity: O(n²) - checking all possible subarrays  
Space Complexity: O(n) - storing prefix sums

```python
class Solution:
    def subarraySum(self, nums: List[int], k: int) -> int:
        result = 0
        nos = len(nums)
        prefix_sum = [0]*(nos+1)

        for i in range(1, nos+1):
            prefix_sum[i] = prefix_sum[i-1] + nums[i-1]
        
        for start in range(nos):
            for end in range(start+1, nos+1):
                if (prefix_sum[end] - prefix_sum[start]) == k:
                    result += 1
        
        return result
```
