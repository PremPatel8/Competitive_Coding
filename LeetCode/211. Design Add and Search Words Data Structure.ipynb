## 211. Design Add and Search Words Data Structure
- Description:
  <blockquote>
  [211. Design Add and Search Words Data Structure](https://leetcode.com/problems/design-add-and-search-words-data-structure/) Design a data structure that supports adding new words and finding if a string matches any previously added string.
 
Implement the `WordDictionary` class:
 
- `WordDictionary()` Initializes the object.
- `void addWord(word)` Adds `word` to the data structure, it can be matched later.
- `bool search(word)` Returns `true` if there is any string in the data structure that matches `word` or `false` otherwise. `word` may contain dots `'.'` where dots can be matched with any letter.
 
**Example:**
**Input**
["WordDictionary","addWord","addWord","addWord","search","search","search","search"]
`[], ["bad"],["dad"],["mad"],["pad"],["bad"],[".ad"],["b.."]`
**Output**
[null,null,null,null,false,true,true,true]
 
**Explanation**
WordDictionary wordDictionary = new WordDictionary();
wordDictionary.addWord("bad");
wordDictionary.addWord("dad");
wordDictionary.addWord("mad");
wordDictionary.search("pad"); // return False
wordDictionary.search("bad"); // return True
wordDictionary.search(".ad"); // return True
wordDictionary.search("b.."); // return True
 
**Constraints:**
 
- `1 <= word.length <= 25`
- `word` in `addWord` consists of lowercase English letters.
- `word` in `search` consist of `'.'` or lowercase English letters.
- There will be at most `2` dots in `word` for `search` queries.
- At most `104` calls will be made to `addWord` and `search`.

### Solution 1, Most Efficient, Trie with Fully Recursive DFS Search

Let M be the length of the word and N be the number of words stored in the trie

Adding a word takes O(M) time since we traverse each character once, and O(M) space in the worst case if all characters create new nodes

For search, it depends on whether the word contains dots.

    With no dots, it's O(M) time — just a straight path down the trie.
    In the worst case with all dots, we branch into every possible child at each level, giving O(26^M) time.
    Space-wise, the recursion stack goes O(M) deep."

"The trie itself takes O(N × M) space to store all the words."

Standard Format

Let:

    M = length of the word (for the current operation)
    N = number of words added

__init__

    Time: O(1)
    Space: O(1)

addWord(word)

    Time: O(M)
    Space: O(M)

search(word)

    Time: O(M) best case, O(26^M) worst case
    Space: O(M) — recursion stack depth

Overall / Trie Storage:

    Space: O(N × M)



```python
class TrieNode:
    def __init__(self) -> None:
        self.children = {}
        self.is_end = False

class WordDictionary:

    def __init__(self):
        self.root = TrieNode()

    def addWord(self, word: str) -> None:
        node = self.root
        
        for ch in word:
            if ch not in node.children:
                node.children[ch] = TrieNode()
            node = node.children[ch]
            
        node.is_end = True

    def _find_node(self, word, node, idx=0):
        if idx == len(word):
            return node.is_end
        
        ch = word[idx]
        
        if ch == ".":
            for child in node.children.values():
                if self._find_node(word, child, idx + 1):
                    return True
            return False
        
        if ch not in node.children:
            return False
        
        return self._find_node(word, node.children[ch], idx + 1)

    def search(self, word: str) -> bool:
        return self._find_node(word, self.root)


# Your WordDictionary object will be instantiated and called as such:
# obj = WordDictionary()
# obj.addWord(word)
# param_2 = obj.search(word)
```

### Solution 2, My Sol Trie with Iterative-Recursive Hybrid Search
Solution description
- Time Complexity: O(N)
- Space Complexity: O(N)

```python
class TrieNode:
    def __init__(self) -> None:
        self.children = {} # ch : TrieNode()
        self.is_end = False

class WordDictionary:

    def __init__(self):
        self.root = TrieNode()

    def addWord(self, word: str) -> None:
        node = self.root
        
        for ch in word:
            if ch not in node.children:
                node.children[ch] = TrieNode()
            
            node = node.children[ch]
        
        node.is_end = True
                

    def _find_node(self, word, node):
        for idx, ch in enumerate(word):
            if ch == ".":
                for child in node.children.values():
                    if self._find_node(word[idx+1:], child):
                        return True
                return False
            elif ch in node.children:
                node = node.children[ch]
            else:
                return False
        
        return node.is_end

    def search(self, word: str) -> bool:
        return self._find_node(word, self.root)
        
        
        
                
        


# Your WordDictionary object will be instantiated and called as such:
# obj = WordDictionary()
# obj.addWord(word)
# param_2 = obj.search(word)
```
